Un taller automotriz necesita llevar un control de su inventario de repuestos. Crea un programa que permita agregar, eliminar y consultar la cantidad de repuestos disponibles, así como realizar una compra cuando el inventario de un repuesto está bajo.

```python
# Almacena los datos
inventario = {}

def agregar_repuesto():

    #Agrega un nuevo repuesto al inventario o actualiza la cantidad si el repuesto ya existe.

    nombre = input("Ingrese el nombre del repuesto: ")
    cantidad = int(input("Ingrese la cantidad: "))
    inventario[nombre] = cantidad
    print(f"Se agregó {cantidad} {nombre}(s) al inventario.")

def eliminar_repuesto():

    #Elimina un repuesto del inventario si existe.

    nombre = input("Ingrese el nombre del repuesto a eliminar: ")
    if nombre in inventario:
        del inventario[nombre]
        print(f"{nombre} fue eliminado del inventario.")
    else:
        print(f"{nombre} no está en el inventario.")

def consultar_repuesto():

    #Muestra la cantidad de un repuesto específico en el inventario.

    nombre = input("Ingrese el nombre del repuesto a consultar: ")
    if nombre in inventario:
        print(f"Hay {inventario[nombre]} {nombre}(s) en el inventario.")
    else:
        print(f"{nombre} no está en el inventario.")

def realizar_compra():

    #Realiza una compra de un repuesto, reduciendo su cantidad en el inventario.

    nombre = input("Ingrese el nombre del repuesto a comprar: ")
    if nombre in inventario:
        cantidad = int(input("Ingrese la cantidad a comprar: "))
        if inventario[nombre] >= cantidad:
            inventario[nombre] -= cantidad
            print(f"Se compraron {cantidad} {nombre}(s).")
        else:
            print(f"No hay suficientes {nombre}(s) en el inventario.")
    else:
        print(f"{nombre} no está en el inventario.")

def mostrar_inventario():

    #Muestra todo el inventario actual, incluyendo nombres de repuestos y cantidades.

    if inventario:
        print("\n--- Inventario Actual ---")
        for repuesto, cantidad in inventario.items():
            print(f"{repuesto}: {cantidad}")
    else:
        print("El inventario está vacío.")

# Menú principal
while True:
    print("\n--- Menú del Inventario ---")
    print("1. Agregar repuesto")
    print("2. Eliminar repuesto")
    print("3. Consultar repuesto")
    print("4. Realizar compra")
    print("5. Mostrar todo el inventario")
    print("6. Salir")

    opcion = input("Elija una opción: ")

    # Ejecuta la función correspondiente según la opción elegida
    if opcion == "1":
        agregar_repuesto()
    elif opcion == "2":
        eliminar_repuesto()
    elif opcion == "3":
        consultar_repuesto()
    elif opcion == "4":
        realizar_compra()
    elif opcion == "5":
        mostrar_inventario()
    elif opcion == "6":
        print("¡Gracias por usar el programa!")
        break
    else:
        print("Opción no válida. Intente de nuevo.")
```

Una empresa de servicios a terceros ofrece proyectos de consultoría y debe llevar el control de horas trabajadas por sus empleados en diferentes proyectos. Crea un programa que calcule las horas totales trabajadas por cada empleado en varios proyectos y su costo total en función de la tarifa por hora.

```python
# Programa para controlar horas trabajadas en proyectos

# Voy a usar diccionarios para guardar la información
empleados = {}
proyectos = {}

# Función para agregar un empleado
def agregar_empleado():
    nombre = input("Ingrese el nombre del empleado: ")
    tarifa = float(input("Ingrese la tarifa por hora del empleado: "))
    empleados[nombre] = {"tarifa": tarifa, "horas_totales": 0}
    print(f"Empleado {nombre} agregado con tarifa de ${tarifa} por hora.")

# Función para agregar un proyecto
def agregar_proyecto():
    nombre = input("Ingrese el nombre del proyecto: ")
    proyectos[nombre] = {}
    print(f"Proyecto {nombre} agregado.")

# Función para registrar horas trabajadas
def registrar_horas():
    empleado = input("Ingrese el nombre del empleado: ")
    proyecto = input("Ingrese el nombre del proyecto: ")
    horas = float(input("Ingrese las horas trabajadas: "))

    if empleado in empleados and proyecto in proyectos:
        if empleado not in proyectos[proyecto]:
            proyectos[proyecto][empleado] = 0
        proyectos[proyecto][empleado] += horas
        empleados[empleado]["horas_totales"] += horas
        print(f"Se registraron {horas} horas para {empleado} en el proyecto {proyecto}.")
    else:
        print("Error: Empleado o proyecto no encontrado.")

# Función para calcular el costo total por empleado
def calcular_costo_total():
    for empleado, datos in empleados.items():
        costo_total = datos["horas_totales"] * datos["tarifa"]
        print(f"{empleado}: {datos['horas_totales']} horas, Costo total: ${costo_total}")

# Función para mostrar horas por proyecto
def mostrar_horas_proyecto():
    for proyecto, datos in proyectos.items():
        print(f"\nProyecto: {proyecto}")
        for empleado, horas in datos.items():
            print(f"  {empleado}: {horas} horas")

# Menú principal
while True:
    print("\n--- Menú de Control de Horas ---")
    print("1. Agregar empleado")
    print("2. Agregar proyecto")
    print("3. Registrar horas trabajadas")
    print("4. Calcular costo total por empleado")
    print("5. Mostrar horas por proyecto")
    print("6. Salir")

    opcion = input("Elija una opción: ")

    if opcion == "1":
        agregar_empleado()
    elif opcion == "2":
        agregar_proyecto()
    elif opcion == "3":
        registrar_horas()
    elif opcion == "4":
        calcular_costo_total()
    elif opcion == "5":
        mostrar_horas_proyecto()
    elif opcion == "6":
        print("¡Gracias por usar el programa!")
        break
    else:
        print("Opción no válida. Intente de nuevo.")
```
